```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.options.display.max_columns=None
```

```python
data=pd.read_csv("HousePrice.csv")
```

<h2>Feature Engineering Summary</h2>
<p>1. Handling missing values</p>
<p>2. Handled Temporal Values</p>
<p>3. Handled Rare Categories in Categorical Features</p>
<p>4. Conversion of all data into a single range to make our model perform better</p>

<h3>Missing Values</h3>

```python
missingfeatures=[i for i in data.columns if data[i].isnull().sum()>0 and data[i].dtype=='O']
for i in missingfeatures:
    print(f'{i} {np.round(data[i].isnull().mean()*100,2)}')
```

```python
df=data.copy()
for i in missingfeatures:
    df[i].fillna('Missing',inplace=True)
df[missingfeatures]
```

```python
#Numerical missing Values
NumericalMissingValues=[i for i in data.columns if data[i].isnull().sum()>0 and data[i].dtype!='O']
NumericalMissingValues
```

```python
pd.options.display.max_rows=10
df[NumericalMissingValues]
```

```python
for i in NumericalMissingValues:
    df[i+' NAN']=np.where(df[i].isnull(),1,0)
    df.fillna(df[i].median(),inplace=True)
df[NumericalMissingValues]
```

<h3>Handling Temporal Variables</h3>

```python
TemporalFeatures=[i for i in data.columns if 'Yr' in i or 'Year' in i]
TemporalFeatures
```

```python
for i in TemporalFeatures:
    if i!='YrSold':
        df[i]=df['YrSold']-df[i]
df[TemporalFeatures].head()
```

```python
df
```

```python
NumericalFeatures=[i for i in data.columns if data[i].dtype!='O' and i not in TemporalFeatures]
data[NumericalFeatures]
```

```python
DiscreteFeatures=[i for i in data.columns if data[i].nunique()<11 and i not in TemporalFeatures+['Id']]
```

```python
ContinuousFeatures=[i for i in NumericalFeatures if i not in TemporalFeatures+DiscreteFeatures+['Id']]
```

```python
#We used log Transformation here
for i in ContinuousFeatures:
    if 0 not in df[i].unique():
        df[i]=np.log(df[i])
df[ContinuousFeatures]
```

<h3>Handling Rare Categories</h3>

```python
CategoricalFeatures=[i for i in data.columns if df[i].dtype=='O']
CategoricalFeatures
```

```python
# If the Categorical Features have category which is available less than 1% than change it to Rare Val
for i in CategoricalFeatures:
    temp=df.groupby(i)['SalePrice'].count()/len(df)
    tempdf=temp[temp<0.01].index
    df[i]=np.where(df[i].isin(tempdf),'Rare Val',df[i])
df[CategoricalFeatures]
```

```python
for feature in CategoricalFeatures:
    labels_ordered=df.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    df[feature]=df[feature].map(labels_ordered)
```

<h3>Feature Scaling</h3>

```python
df.head()
```

```python
feature_scale=[i for i in data.columns if i not in ['Id','SalePrice']]
from sklearn.preprocessing import MinMaxScaler
scale=MinMaxScaler()
scale.fit(df[feature_scale])
```

```python
scale.transform(df[feature_scale])
```

```python
dataset=pd.concat([df[['Id','SalePrice']].reset_index(drop=True),pd.DataFrame(scale.transform(df[feature_scale]),columns=feature_scale)],axis=1)
dataset
```

```python
dataset.to_csv("HousePriceFinal.csv",index=False)
```
